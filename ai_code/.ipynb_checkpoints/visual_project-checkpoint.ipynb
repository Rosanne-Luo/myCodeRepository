## 获取测试数据集

```python
import pandas as pd

def get_imbalanced_datasets():
    """ 获取不均衡数据集样例
    
    Args:
        None
        
    Return:
        data_all: 带标签的数据集
    """
    alexa_data = pd.read_csv("./data/top-1000000-domains.txt", names=["domain"])
    dga_data = pd.read_csv("./data/feodo.txt", sep="\t", skiprows=17, names=["domain", "time1", "time2"])
    dga_data.drop(["time1","time2"], axis=1, inplace=True)
    dga_data["label"] = 1

    alexa_data = alexa_data.iloc[0:100000,:]
    alexa_data["label"] = 0
    data_all = pd.concat([alexa_data, dga_data])
    return data_all

from sklearn import feature_extraction
from sklearn.model_selection import train_test_split

def feature_unigram(data_all):
    """ 通过unigram的方法特征化data_all
    
    Args:
        data_all: N*1 矩阵，每一行为一个字符串
    
    Returns:
        X : 特征向量
    
    """
    X = data_all["domain"]
    y = data_all["label"]
    ngram_vectorizer = feature_extraction.text.CountVectorizer(analyzer='char', ngram_range=(1, 1))
    X = ngram_vectorizer.fit_transform(X).toarray()
    return X, y

```

```python
data_all = get_imbalanced_datasets()
X, y = feature_unigram(data_all)
```

## 可视化源数据

## PCA 降维可视化

```python
# 可视化源数据
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
import seaborn as sns

def visual_dataset_PCA(X, y):
    """ 可视化原始数据集（二分类）
    
    Args:
        X: 已经经过处理的数字化的特征向量 
        y:  特征向量对应的0-标签
        
    Returns:
        Null. 产生类别散点图和柱状图
    
    """
    pca = PCA(n_components = 2)
    pca.fit(X)
    X_reduced = pca.transform(X)
    
    fig, ax = plt.subplots(1, 2, figsize=(12,5))
    
    ax[0].scatter(X_reduced[y == 0, 0], X_reduced[y == 0, 1], label = "alexa domains", alpha = 0.2)
    ax[0].scatter(X_reduced[y == 1, 0], X_reduced[y == 1, 1], label = "dga data", alpha = 0.2)
    ax[0].set_title("PCA of orginal dataset")
    ax[0].legend()
    
    ax[1] = sns.countplot(y)
    ax[1].set_title("Number of observations per class")
```

```python
visual_dataset_PCA(X, y)
```
